```python
with open("data/raw/ActivityNet/val_2_video_status.csv", "w") as f:
    f.write("video_id,is_viewable\n")
    for video_id in list(data.keys()):
        video_id = video_id[2:]
        url = f'https://www.googleapis.com/youtube/v3/videos?id={video_id}&key={api_key}&part=status'
        url_get = requests.get(url)
        try:
            is_viewable = url_get.json()["items"][0]["status"]["publicStatsViewable"]
        except:
            is_viewable = False
        f.write(f"{video_id},{is_viewable}\n")
        print(f"{video_id}: {is_viewable}")
```
